### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#Earlier I did not see the details for penang Flavors so I added it with basic info. Here, I am updating the document to perform the search
updated_data = {
    "BusinessName": "Penang Flavours",
    "BusinessType": "Restaurant/Cafe/Canteen",
    "BusinessTypeID": 1,
    "AddressLine1": "Penang Flavours",
    "AddressLine2": "146A Plumstead Rd",
    "AddressLine3": "London",
    "AddressLine4": "",
    "PostCode": "SE18 7DY",
    "Phone": "",
    "LocalAuthorityCode": "511",
    "LocalAuthorityName": "Greenwich",
    "LocalAuthorityWebSite": "http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress": "health@royalgreenwich.gov.uk",
    "scores": {
        "Hygiene": None,
        "Structural": None,
        "ConfidenceInManagement": None
    },
    "SchemeType": "FHRS",
    "geocode": {
        "longitude": 0.08384000,
        "latitude": 51.49014200
    },
    "RightToReply": "",
    "Distance": 4623.9723280747176,
    "NewRatingPending": True
}

# Update the document
db.establishments.update_one(
    {"BusinessName": "Penang Flavours"},
    {"$set": updated_data}
)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
penang_flavours = db.establishments.find_one({"BusinessName": "Penang Flavours"}, {"geocode.latitude": 1, "geocode.longitude": 1, "_id": 0})
latitude = penang_flavours["geocode"]["latitude"]
longitude = penang_flavours["geocode"]["longitude"]

query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

results = db.establishments.find(query).sort("scores.Hygiene", 1).limit(5)


# Print the results
for doc in results:
    pprint(doc)
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.DataFrame(list(db.establishments.find(query).sort("scores.Hygiene", 1).limit(5)))
print(f"Total rows: {len(top5_df)}")
top5_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [
    {"$match": {"scores.Hygiene": 0}},  
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},  
    {"$sort": {"count": -1}}  
]

hygiene_results = list(db.establishments.aggregate(pipeline))

# Print the total number of Local Authorities in the result
print(f"Total Local Authorities with hygiene score 0: {len(hygiene_results)}")

# Print the first 10 results
pprint(hygiene_results[:10])
```

```python
#I think while entering the database again there were duplicate entries so debugging for that (since all results are double of sample)

from collections import Counter 

all_businesses = list(db.establishments.find({}, {"BusinessName": 1, "AddressLine1": 1, "_id": 0}))

business_list = [(entry.get("BusinessName", "Unknown"), entry.get("AddressLine1", "Unknown")) for entry in all_businesses]

duplicates = [item for item, count in Counter(business_list).items() if count > 1]

print(f"Total duplicate businesses found: {len(duplicates)}")
pprint(duplicates[:10])  
```

```python
pipeline = [
    {"$group": {
        "_id": {"BusinessName": "$BusinessName", "AddressLine1": "$AddressLine1"},
        "ids": {"$push": "$_id"},  # Store all document IDs
        "count": {"$sum": 1}
    }},
    {"$match": {"count": {"$gt": 1}}}  # Only process duplicates
]

duplicates = list(db.establishments.aggregate(pipeline))

print(f"Total duplicate groups found: {len(duplicates)}")
```

```python
for record in duplicates:
    ids_to_remove = record["ids"][1:]  # Keep the first entry, remove the rest
    db.establishments.delete_many({"_id": {"$in": ids_to_remove}})

print("Duplicate entries removed.")
```

```python
recheck_duplicates = list(db.establishments.aggregate(pipeline))

print(f"Remaining duplicate groups: {len(recheck_duplicates)}")
```
